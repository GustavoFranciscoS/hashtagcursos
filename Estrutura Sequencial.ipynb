#### 3. Faça um Programa que peça dois números e imprima a soma.

```python
numero1 = int(input('Qual o primeiro numero?'))
numero2 = int(input('Qual o segundo numero?'))

print(numero1)
print(numero2)

soma = numero1 + numero2

print(f'A soma é {soma}')
```

#### 4. Faça um Programa que peça as 4 notas bimestrais de um aluno e mostre a média de todas as notas.

```python
nota1 = float(input('Qual a primeira nota?'))
nota2 = float(input('Qual a segunda nota?'))
nota3 = float(input('Qual a terceira nota?'))
nota4 = float(input('Qual a quarta nota?'))

print(f'A primeira nota foi {nota1}')
print(f'A segunda nota foi {nota2}')
print(f'A terceira nota foi {nota3}')
print(f'A quarta nota foi {nota4}')

media = (nota1 + nota2 + nota3 + nota4) / 4

print(f'A média do bimestre foi: {media}')
```

#### 5. Faça um Programa que converta metros para centímetros. Você pode pedir o comprimento em metros para o usuário (input).

```python
metros = int(input('Informe a quantidade em metros:'))

print(metros)

centimetros = metros * 100

print(f'Aquantidade de {metros} metros, em centimetros é: {centimetros} centimetros')
```

#### 6. Faça um Programa que calcule a área de uma sala de um apartamento. Para isso, o seu programa precisa pedir a largura da sala, o comprimento da sala e imprimir a área em m² da sala.

```python
largura = float(input('Qual a largura da sala?'))
print(f'A largura da sala é {largura}')

comprimento = float(input('Qual o comprimento da sala?'))
print(f'O comprimento da sala é {comprimento}')

metros_quadrados = largura * comprimento
print(f'A sala possui {metros_quadrados} metros quadrados')
```
